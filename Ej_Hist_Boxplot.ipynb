```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#Estilos
from matplotlib import style
#ref: https://www.geeksforgeeks.org/style-plots-using-matplotlib/
from matplotlib.patches import Rectangle
#necesario para dibujar leyenda

#CARGA DE DATOS
filename='/content/catsvdogs.xlsx'
catdogdf=pd.read_excel(filename)
catdogdf.head()
```

<br> 1. Trazo un histograma del porcentaje de dueños de perros.

---



```python
x1=catdogdf['Percentage of Dog Owners']

#FORMATO DE MATPLOTLIB.PLOT SE UTILIZO LA SIGUIENTE REFERENCIA
#ref https://matplotlib.org/stable/tutorials/introductory/pyplot.html

plt.title('Histograma % dueños de perros',fontsize=15)
plt.xlabel('% dueños de perros')
plt.ylabel('Frecuencia / conteo',fontsize=12)

plt.hist(x1, bins=20, ec='black') #ec para color lineas en barras
plt.style.use('fast') #estilo fast de matplotlib
```

<br> 2. Trazo un histograma del porcentaje de dueños de gatos.

---



```python
x2=catdogdf['Percentage of Cat Owners']

#FORMATO DE MATPLOTLIB.PLOT SE UTILIZO LA SIGUIENTE REFERENCIA
#ref https://matplotlib.org/stable/tutorials/introductory/pyplot.html

plt.title('Histograma % dueños de gatos',fontsize=15)
plt.xlabel('% dueños de gatos')
plt.ylabel('Frecuencia / conteo',fontsize=12)

plt.hist(x2, bins=20, color='yellow', ec='black')
#ec para color lineas en barras
#color para color de barras

plt.style.use('fast') #estilo fast de matplotlib
```

<br> 3a. ¿Qué podemos ver al comparar estos dos histogramas? ¿Qué información

---

nos dice?

<i>Nos dice que, en la mayoría de estados, hay más porcentaje de casas con perros que con gatos.  Vemos que en promedio, los estados tienen alrededor de 30% de casas con gatos.  Mientras que, en promedio, los estados parecen tener alrededor del 35-40% de casas con perros. </i>

<br> 3b.  <b> Opcional/bono </b>: Los dos histogramas en el mismo trazado.

---



```python
plt.title('Histograma % dueños de gatos vs % dueños de perros',fontsize=15)
plt.xlabel('% de dueños')
plt.ylabel('Frecuencia / conteo',fontsize=12)


pltgatos=plt.hist(x1, bins=20, ec='black')
pltperros=plt.hist(x2, bins=20, color='yellow', alpha=0.8, ec='black')
#alterar la propiedad alpaha para transparencia
#https://www.geeksforgeeks.org/change-the-line-opacity-in-matplotlib/#:~:text=Matplotlib%20allows%20you%20to%20regulate,such%20as%200.5%20or%200.25.

#PARA CREAR LEYENDA SE ESTABLECE UN COLOR PARA CADA HISTOGRAMA
#https://stackoverflow.com/questions/43872450/matplotlib-histogram-with-multiple-legend-entries
colg='yellow'
colp='blue'
#SE GRABA UN RECTANGULO AMARILLO Y OTRO AZUL EN handles
handles = [Rectangle((0,0),1,1,color=c,ec="k") for c in [colg,colp]]
#EN labels SE LLAMA A CADA NOMBRE DE HISTOGRAMA
labels= ["% gatos","% perros"]
plt.legend(handles, labels, loc='upper right',fontsize=15)

plt.style.use('fast') #estilo fast de matplotlib
```

En general, el cúmulo de casas con perros está más a la derecha que el cúmulo de casas con gatos, es decir, en general los estados tienen más porcentaje de casas con perros que con gatos.

<br> 4. Creo dos boxplots en mismo trazado: uno en la media del número de gatos por hogar y el otro de la media del número de perros.

---



```python
x1=catdogdf['Mean Number of Cats']
x2=catdogdf['Mean Number of Dogs per household']
fig, axes = plt.subplots(nrows = 1, ncols = 1)
axes.boxplot([x1,x2], notch = True, labels=['Gatos', 'Perros']);
plt.ylabel('PROMEDIO DE ANIMALES POR CASA EN LOS ESTADOS')
```

<br> 5. ¿Qué podemos ver al comparar estos dos boxplots? ¿Qué información nos dice?

---



*Vemos que la media de numero de gatos por casa es mayor que la de perros por casa.  Esto es, por lo general, las casas que tienen una de las dos mascotas tienen en promedio, mas numero de gatos que de perros.*
